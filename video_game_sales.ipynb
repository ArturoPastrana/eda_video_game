## Filtros con condicionales logicos usando `loc[ ]`

```python
# Forma de ver los valores unicos de cada columna
for i in data[['Name','Platform','Year','Genre','Publisher']]:
    print(i.upper())
    print(data[i].unique())
    print('')
```

```python
nes = data['Platform'] == 'NES'
```

```python
nes
```

```python
nes.value_counts()
```

```python
data.loc[nes]
```

```python
data.loc[data['Platform'] == 'GB']
```

```python
data.loc[data['NA_Sales'] > data['JP_Sales'],'Name']
```

```python
# Para este caso ambas condiciones deben cumplirse 
data.loc[(data['Platform'] == 'GB') & (data['Genre'] == 'Role-Playing')]
```

```python
# En este caso si se cumple una de las condiciones la incluye como cuando ambas se cumplen
data.loc[(data['Platform'] == 'GB') | (data['Genre'] == 'Role-Playing')]
```

```python
data.loc[(data['Platform'] == 'GB') | (data['Genre'] == 'Role-Playing')].sort_values(by = 'Global Sales', ascending = False)
```
